```r
library(dplyr)
library(lubridate)
library('ggplot2') 
library('ggthemes')
library(grid)
library(RColorBrewer)
library(gplots)
library(forecast)
library(gridExtra)
```

```r
# Load Arctic Sea Ice Extent data
# Source: ftp://sidads.colorado.edu/DATASETS/NOAA/G02135/north/daily/data/N_seaice_extent_daily_v3.0.csv
# Downloaded on 22nd May 2020 (last day of the dataset)
seaice_N = read.csv("./data/N_seaice_extent_daily_v3.0.csv", stringsAsFactors=FALSE)

# Have a look at the table
head(seaice_N) # tail()
```

```r
# The first row of each column doesn't contain data, so remove it
seaice_N = seaice_N[-c(1),]

# Convert factors into integers
seaice_N$Year = as.integer(seaice_N$Year)
seaice_N$Month = as.integer(seaice_N$Month)
seaice_N$Day = as.integer(seaice_N$Day)

# Convert the Sea Ice Extent into numeric class
seaice_N$Extent = as.numeric(seaice_N$Extent)
seaice_N$Missing = as.numeric(seaice_N$Missing)

seaice_N = seaice_N %>%
  select(-Source.Data, -Missing) %>% # Remove Source.Data and Missing columns
  mutate(Date = ymd(paste(Year, Month, Day, sep = "-"))) # Create Date column

ice = seaice_N %>%
  group_by(Year,Month) %>%
  summarize(avg_extent=mean(Extent)) 

# Make a vector with all months
ice_series = as.vector(unlist(ice[1:500,3]))
ice_series = ts(ice_series, start = c(1978, 10), end = c(2020, 5), frequency = 12)
```

```r
# Boxplot
boxplot(ice_series~cycle(ice_series), xlab = "Month", ylab ="Million sq km",
                         main ='Arctic Sea Ice Extent')
```

```r
# Decompose
decomposed_seaice = decompose(ice_series, type = "multiplicative")

# Decomposition plot
options(repr.plot.width=15, repr.plot.height=10)
ndec = forecast::autoplot(decomposed_seaice) + 
  ggtitle(paste('Arctic Sea Ice Extent')) +
  scale_y_continuous(position = 'right') + 
  theme(text = element_text(size=20))
grid.arrange(ndec, ncol = 1)
```

```r
# Build the model:
# Fit the model. We'll use from 1978-11 until 2017-12, to have the last years for validation
# mymodel = auto.arima(ice_series[1:471], trace = TRUE)

mymodel = auto.arima(ice_series[1:471], lambda=0, d=0, D=1,
         max.order=4, stepwise=FALSE, approximation=FALSE)
```

```r
# Plotting the residuals allow to choose the model:
# pacf(ts(mymodel$residuals), main = 'PACF Residual')
checkresiduals(mymodel)

# The ACF doesn't look like white noise, because not 
# all autocorrelations are within the threshold limits
# This indicates we have to choose another model
```

```r
# Make the forecast
myforecast = forecast(mymodel, level = c(95), h = 29) # Forecast 29 months
plot(myforecast)
```

```r
# Test the accuracy of the model for different lags
Box.test(mymodel$resid, lag = 2, type = "Ljung-Box")
Box.test(mymodel$resid, lag = 10, type = "Ljung-Box")
Box.test(mymodel$resid, lag = 50, type = "Ljung-Box")
# p-values are low, but x-squared high
```
